```python
import copy
import json
import codecs
import jieba
from collections import defaultdict
from dateutil.parser import parse
flatten = lambda x: [y for l in x for y in flatten(l)] if type(x) is list else [x]
type_ref_dict = {
    'Action':'动作',
    'Adventure':'冒险',
    'Animation':'动画',
    'Biography':'传记',
    'Comedy':'喜剧',
    'Crime':'犯罪',
    'Drama':'戏剧',
    'Family':'家庭',
    'Fantasy':'奇幻',
    'History':'历史',
    'Horror':'惊悚',
    'Music':'音乐',
    'Musical':'歌舞',
    'Mystery':'悬疑',
    'Romance':'爱情',
    'Sci-Fi':None,
    'Short':None,
    'Sport':'运动',
    'Thriller':'恐怖',
    'War':'战争',
}
def parse_ch_duration(duration):
    try:
        return int(duration.split('分钟')[0].strip())
    except Exception as e:
        pass
def parse_en_duration(duration):
    temp = duration.split()
    if len(temp)==2:
        return int(temp[0].split('h')[0])*60+int(temp[1].split('min')[0])
    else:
        if temp[0].endswith('h'):
            return int(temp[0].split('h')[0].strip())*60
        if temp[0].endswith('min'):
            return int(temp[0].split('min')[0].strip())
def parse_ch_releaseDate(releaseDate):
    res = []
    for i in releaseDate:
        res.append(parse(i.split('(')[0].strip()))
    return res
def parse_en_releaseDate(releaseDate):
    try:
        return parse(releaseDate.split('(')[0].strip())
    except Exception as e:
        pass
def cal_jaccard_sim(en_info_for_jaccard, ch_info_for_jaccard):
    en_copy = copy.deepcopy(en_info_for_jaccard)
    ch_copy = copy.deepcopy(ch_info_for_jaccard)
#     print(id(en_copy))
#     print(id(en_info_for_jaccard))
    sim_dict = defaultdict(list)
    for en_key,en_value in en_copy.items():
        temp = []
        map_list = []
        for i in en_value[0]:
            map_list.append(type_ref_dict[i])
        en_value[0] = map_list
        for ch_key,ch_value in ch_copy.items():
            temp.append(tuple([len(set(flatten(en_value)) & set(flatten(ch_value)))/len(set(flatten(en_value)) | set(flatten(ch_value))),ch_key]))
#         sim_dict[en_key] = temp
        temp = sorted(temp,key=lambda x:x[0],reverse=True)[0]
        sim_dict[en_key] = temp
    return sim_dict
#返回前k个
def top_k_jaccard_sim(en_info_for_jaccard, ch_info_for_jaccard, k=None):
    en_copy = copy.deepcopy(en_info_for_jaccard)
    ch_copy = copy.deepcopy(ch_info_for_jaccard)
    sim_dict = defaultdict(list)
    for en_key,en_value in en_copy.items():
        temp = []
        map_list = []
        for i in en_value[0]:
            map_list.append(type_ref_dict[i])
        en_value[0] = map_list
        for ch_key,ch_value in ch_copy.items():
            temp.append(tuple([len(set(flatten(en_value)) & set(flatten(ch_value)))/len(set(flatten(en_value)) | set(flatten(ch_value))),ch_key]))
#         sim_dict[en_key] = temp
        if k:
            temp = sorted(temp,key=lambda x:x[0],reverse=True)[:k]
        else:
            temp = sorted(temp,key=lambda x:x[0],reverse=True)
        sim_dict[en_key] = temp
    return sim_dict
```

```python
res = []
ch_info_for_jaccard = defaultdict(list)
en_info_for_jaccard = defaultdict(list)
ch_brief = {}
en_brief = {}
with codecs.open('movies.jl','r',encoding='utf-8') as f:
    for i in f:
        res.append(json.loads(i))
count = 0
for i in res:
    temp = None
    i['movie_id'] = count
    count += 1
    temp = [ i['type'], i['duration'], i['release_date'], i['ratings'] ]
    if int(i['english']) == 0: 
        temp[1] = parse_ch_duration(temp[1])
        temp[2] = parse_ch_releaseDate(temp[2])
        ch_info_for_jaccard[i['movie_id']] = temp
        brief_str = ''.join(i['brief_into'].split())
        seg_generator = jieba.cut(brief_str)
        ch_brief[i['movie_id']] = ' '.join(seg_generator)
    if int(i['english']) == 1:
        temp[1] = parse_en_duration(temp[1])
        temp[2] = parse_en_releaseDate(temp[2])
        en_info_for_jaccard[i['movie_id']] = temp
        en_brief[i['movie_id']] = i['brief_into']
```

```python
def hit_10_3(top_10_sim_dict,top_3_sim_dict,labels,threshold):
    count_10 = 0
    count_3 = 0
    top_10_copy = copy.deepcopy(top_10_sim_dict)
    top_3_copy = copy.deepcopy(top_3_sim_dict)
    top_10_preds = []
    top_3_preds = []
    for key,value in top_10_copy.items():
        temp = []
        for i in value:
            if i[0] >= threshold:
                temp.append(set([key,i[1]]))
        top_10_preds.extend(temp)
    for key,value in top_3_copy.items():
        temp = []
        for i in value:
            if i[0] >= threshold:
                temp.append(set([key,i[1]]))
        top_3_preds.extend(temp)
#     print(top_3_preds[:10])
#     print(top_10_preds[:10])
    for i in labels:
        if i in top_10_preds:
            count_10 += 1
        if i in top_3_preds:
            count_3 +=1
    return [tuple([10*count_10/len(top_10_preds),count_10/len(labels)]),tuple([3*count_3/len(top_3_preds),count_3/len(labels)])]
```

```python
def jaccard_eval(preds_dict, labels, threshold):
    count = 0
    preds_list = []
    for key,value in preds_dict.items():
        if value[0] >= threshold:
            preds_list.append(set([value[1],key]))
    for i in preds_list:
        if i in labels:
            count += 1
    return ((count+1)/(len(preds_list)+1),count/len(labels))   
```

```python
import numpy as np
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer

n_features = 20
n_topics = 10
stop_words_list = [ i.strip() for i in codecs.open('stop_words_ch.txt','r',encoding='utf-8')]
stop_words_list.extend([i.strip() for i in codecs.open('stop_words_en.txt','r',encoding='utf-8')])
index2ch_movie_id = {key:value for key,value in enumerate(ch_brief.keys())}
index2en_movie_id = {key:value for key,value in enumerate(en_brief.keys())}
#获得主题相似度矩阵
def topics_sims(model, ch_brief, en_brief, model_tfidf_vectorizer):
    topics_num = model.components_.shape[0]
    feature_num = model.components_.shape[1]
    print('topics_num:%d, feature_num:%d' % (topics_num, feature_num))
    m = len(en_brief.keys())
    n = len(ch_brief.keys())
#     print('m:%d, n:%d' % (m,n))
    en_brief_list = list(en_brief.values())
    ch_brief_list = list(ch_brief.values())
    en_topics = np.ones((m,topics_num))
    ch_topics = np.ones((n,topics_num))
    m_index = 0
    n_index = 0
    for i,j in enumerate(en_brief_list):
        en_brief_tfidf = model_tfidf_vectorizer.fit_transform([j])
        try:
            #存在样本维度过小达不到max_feature
            en_topics[i,:] = model.transform(en_brief_tfidf)
        except Exception as e:
            pass
    for p,q in enumerate(ch_brief_list):
        ch_brief_tfidf = model_tfidf_vectorizer.fit_transform([q])
        try:
            #存在样本维度过小达不到max_feature
            ch_topics[p,:] = model.transform(ch_brief_tfidf)
        except Exception as e:
            pass
    res = en_topics.dot(ch_topics.T)
    for k in range(m):
        for l in range(n):
            res[k,l] = res[k,l]/(np.sqrt(np.sum(en_topics[k,:]**2))*np.sqrt(np.sum(ch_topics[l,:]**2)))
    return res
#lda预测
def lda_predict(topics_similarites, ch_index_dict, en_index_dict):
    mn_index = []
    preds_list = []
    for i in range(len(topics_similarites.max(axis=1))):
        index = np.where(topics_similarites[i,:]==topics_similarites.max(axis=1)[i])
        mn_index.append(index[0][0])
    for k,l in enumerate(mn_index):
        preds_list.append(set([en_index_dict[k],ch_index_dict[l]]))
    return preds_list
#准确率，召回率
def lda_eval(lda_preds, labels):
    count = 0
    for i in lda_preds:
        if i in labels:
            count += 1
    return (count/len(lda_preds),count/len(labels))
#只用lda时，进行参数搜索，但
def search_params(n_features, n_topics):
    lda_tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words_list,max_features=n_features) 
    lda_tfidf = lda_tfidf_vectorizer.fit_transform(documents)
    lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
    lda.fit(lda_tfidf)
    topics_similarites = topics_sims(lda, ch_brief, en_brief, lda_tfidf_vectorizer)
    lda_preds = lda_predict(topics_similarites, ch_index_dict, en_index_dict)
    P,R = lda_eval(lda_preds, labels)
    print('P:%f, R:%f' % (P,R))
documents = []
documents = list(ch_brief.values())
documents.extend(list(en_brief.values()))

lda_tfidf_vectorizer = TfidfVectorizer(stop_words=stop_words_list,max_features=n_features) 
lda_tfidf = lda_tfidf_vectorizer.fit_transform(documents)
lda = LatentDirichletAllocation(n_topics=n_topics, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
lda.fit(lda_tfidf)
topics_similarites = topics_sims(lda, ch_brief, en_brief, lda_tfidf_vectorizer)
#只用lda效果不好
lda_preds = lda_predict(topics_similarites, index2ch_movie_id, index2en_movie_id)
P,R = lda_eval(lda_preds, labels)
print('P:%f, R:%f' % (P,R))
# for n_features in range(5,30):
#     for n_topics in range(2,15):
#         search_params(n_features, n_topics)
```
